### Check if date X is between date A and date B
Assuming dates x, a, b are all in string format, e.g. "2011-01-01"

```python
# from datetime import datetime
def is_date_between(date_to_check, start_date, end_date):
    date_format = '%Y-%m-%d'
    d2 = datetime.strptime(start_date, date_format)
    d1 = datetime.strptime(date_to_check, date_format)
    d3 = datetime.strptime(end_date, date_format)
    if d2 < d1 < d3:
        return 'yes'
    else:
        return 'no'
```

```python
date_x = '2015-06-01'
date_a = '2014-08-22' # before date_x
date_b = '2015-07-05' #  after date_x
date_c = '2015-02-19' # before date_x

# Should equal yes
is_date_between(start_date=date_a, date_to_check=date_x, end_date=date_b)
# Should equal no
is_date_between(start_date=date_a, date_to_check=date_x, end_date=date_c)
```
